```python
import pandas as pd
dataset_Ayes = pd.read_excel("general_data1.xlsx",sheet_name=0)
dataset_Ano = pd.read_excel("general_data1.xlsx",sheet_name=1)
d1 = dataset_Ayes.head(20)
d2 = dataset_Ano.head(20)
d1
```

```python
d2
```

# 1. Mannwhitney Test

```python
from scipy.stats import mannwhitneyu
stats,p = mannwhitneyu(d1.PercentSalaryHike,d2.PercentSalaryHike)
print("stats: ",stats,'\n',"P value: ",p)
```

```python
stats,p = mannwhitneyu(d1.Education,d2.Education)
print("stats: ",stats,'\n',"P value: ",p)
```

# 2. Wilcoxon Test

```python
from scipy.stats import wilcoxon
stats,p = wilcoxon(d1.Age,d1.Education)
print("stats: ",stats,'\n',"P value: ",p)
```

# 3. Friedman Test

```python
from scipy.stats import friedmanchisquare
stats,p = friedmanchisquare(d1.Age,d1.PercentSalaryHike,d1.Education)
print("stats: ",stats,'\n',"P value: ",p)
```

# 4. Kruskal Wallis Test

```python
from scipy.stats import kruskal
stats,p = kruskal(d2.Age,d2.Education,d2.PercentSalaryHike)
print("stats: ",stats,'\n',"P value: ",p)
```

# 6. One Sample Test

```python
from scipy.stats import ttest_1samp
a1 = data4.PercentSalaryHike
stat,p = ttest_1samp(a1,5)
print("stats: ",stats,'\n',"P value: ",p)
```

# 7. Two Sample Test

```python
from scipy.stats import ttest_rel
a2 = data4.Age
a3 = data4.PercentSalaryHike
stat,p = ttest_rel(a2,a3)
print("stats: ",stats,'\n',"P value: ",p)
```

# 8.Two Sample Independent T-Test

```python
from scipy.stats import ttest_ind
a2 = data4.Age
a3 = data4.PercentSalaryHike
stat,p = ttest_rel(a2,a3)
print("stats: ",stats,'\n',"P value: ",p)
```
