```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import neighbors
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("train.csv")
dataset.head()
```

```python
le = preprocessing.LabelEncoder()
le.fit(dataset["Sex"])
dataset["Sex"] = le.transform(dataset["Sex"])
le.fit(dataset["Embarked"])
dataset["Embarked"] = le.transform(dataset["Embarked"])
dataset.head()
```

```python
y  = dataset["Pclass"]
X = dataset.drop(["Pclass","PassengerId","Name","Ticket","Cabin"],axis=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
def K_value(k):
    knn=neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train).score(X_test,y_test)
    y_pred=knn.predict(X_test)
    print("\nConfusion matrix when k = ",k,":\n",confusion_matrix(y_test,y_pred))
    print("Accuracy is when k = ",k,":",accuracy_score(y_test,y_pred,normalize=True))
    
for i in range(1,268):
    K_value(i)
```

# From above Observation k = 1 gives more accuracy than other values.
