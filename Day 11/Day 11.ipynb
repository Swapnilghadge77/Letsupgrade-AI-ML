```python
import pandas as pd
import matplotlib.pyplot as plt
dataset2 = pd.read_csv("general_data.csv")
dataset2["Attrition"] = dataset2["Attrition"].map({"Yes":1,"No":0})
dataset2["Gender"] = dataset2["Gender"].map({'Male':1, 'Female':0})
dataset2.head(20)
```

```python
dataset2.isnull
```

```python
dataset2.dropna
```

```python
dataset2.describe
```

# 1. Correlation of Attrition & Age

```python
from scipy.stats import pearsonr
stats,p = pearsonr(dataset2.Attrition,dataset2.Age)
print(stats,p)
if p<0.05:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
plt.scatter(dataset2.Attrition,dataset2.Age)
```

```python
dataset2.corr()
```
