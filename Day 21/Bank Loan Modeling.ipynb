# Logistic Regression for Bank Loan Modeling Problem 

```python
# Logistic Regression for Bank Loan Modeling Problem 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
df.head()
```

```python
df.isna()
```

```python
df.dropna()
```

```python
df.describe()
```

```python
y = df[["Personal Loan"]]
x = df[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','Online','CreditCard']]
```

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(x)
Logistics = sm.Logit(y,x1)
result = Logistics.fit()
result.summary()
```

# Conclusion
From the Summary 'Income', 'Family', 'CCAvg', 'Education', 'Securities Account' are significant because both of their p-values are "<0.05".
